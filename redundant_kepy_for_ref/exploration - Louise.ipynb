## lesion_id

There are duplicates for the lesion ids. Let us verify that each lesion corresponds to 1 individual

```python
lesion_id_duplicates = info.groupby('lesion_id').count()['image_id']
lesion_id_duplicates
```

```python
lesion_id_duplicates.describe()
```

```python
lesion_id_duplicates = pd.DataFrame(lesion_id_duplicates)
lesion_id_duplicates = lesion_id_duplicates[lesion_id_duplicates.image_id > 1]
lesion_id_duplicates.head()
```

```python
lesion_id_duplicates['age_min'] = info.groupby('lesion_id').min()['age']
lesion_id_duplicates['age_max'] = info.groupby('lesion_id').max()['age']
```

```python
lesion_id_duplicates.head()
```

```python
def equal_age(age) : 
    return (age[0] == age[1])
```

```python
lesion_id_duplicates['same_age'] = lesion_id_duplicates[['age_min', 'age_max']].apply(lambda x: equal_age(x), axis = 1)
lesion_id_duplicates.head()
```

```python
lesion_id_duplicates[lesion_id_duplicates.same_age == 0]
```

Ages match, except for when age is missing
